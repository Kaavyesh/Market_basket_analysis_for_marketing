Association Rules

```python
transaction = []
for item in df['Transaction'].unique():
  lst = list(set(df[df['Transaction'] == item]['Item']))
  transaction.append(lst)
```

```python
transaction[0:10]
```

```python
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
from mlxtend.preprocessing import TransactionEncoder
```

```python
te = TransactionEncoder()
encodeddata = te.fit(transaction).transform(transaction)
basket_dataset = pd.DataFrame(encodeddata,columns = te.columns_)
basket_dataset
```

```python
basket_dataset.shape
```

Apriori algorithm :
why:Apriori is a popular algorithm for extracting frequent itemsets with applications in association rule learning.  An itemset is considered as "frequent" if it meets a user-specified support threshold.

```python
frequent_item = apriori(basket_dataset,min_support = 0.01,use_colnames = True)
frequent_item.sort_values('support',ascending = False)
```

bare Coffee is chosen in 48% of cases. Coffee and Bread - in 9%. Bare salad - in 1%. This list is a necessary list for creating association rules using the function association_rules() that generates a DataFrame of association rules including the metrics scores.

Rule generation is a common task in the mining of frequent patterns.

```python
rules = association_rules(frequent_item, metric="confidence", min_threshold=0.2)
rules.sort_values('confidence', ascending = False, inplace=True)
rules
```

1. **Antecedents**: Represents the items or itemsets in the "if" part of an association rule, indicating what is considered in the transaction.
2. **Consequents**: Represents the items or itemsets in the "then" part of an association rule, indicating what is frequently associated with the antecedents.
3. **Antecedent Support**: Measures how often the antecedents (items in the "if" part) appear in transactions.
4. **Consequent Support**: Measures how often the consequents (items in the "then" part) appear in transactions.
5. **Support**: Measures the frequency of the combined itemset of both antecedents and consequents occurring together.
6. **Confidence**: Measures the likelihood of seeing the consequents in a transaction given the presence of the antecedents.
7. **Lift**: Measures how much more often the antecedents and consequents appear together than expected if they were independent.
8. **Leverage**: Computes the difference between observed and expected co-occurrence of antecedents and consequents.
9. **Conviction**: Indicates how much the consequents depend on the antecedents.
10. **Zhang's Metric**: A metric used in association rule analysis to assess the interestingness of rules based on support and confidence.

```python
rules[rules["lift"]>1].sort_values("support",ascending = False)
```
